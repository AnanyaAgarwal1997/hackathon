```python
def extract_DOJ(df):
    df['Date_Crime'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p')
    df['Crime_Day'] = df['Date_Crime'].dt.day
    df['Crime_Month'] = df['Date_Crime'].dt.month
    df['Crime_Year'] = df['Date_Crime'].dt.year
    df = df.drop('Date', axis=1)
    return df

df = extract_DOJ(df)
df
```

```python
def crime_Time(df):
    df['Time'] = pd.to_datetime(df['Time'], format='%H:%M')
    df['Dep_Hour'] = df['Time'].dt.hour
    df['Dep_Minute'] = df['Time'].dt.minute
    df = df.drop('Time', axis=1)
    return df

df = crime_Time(df)
df
```

```python
df_binary = pd.get_dummies(df['DayOfWeek'], prefix='Day', dtype=int)
df_binary
```

```python
df = pd.concat([df, df_binary], axis = 1)
df = df.drop('DayOfWeek', axis=1)
df
```

```python
import geopandas as gpd
from shapely.geometry import Point
from geopy.distance import distance

# Load parking lots and restaurants data into GeoDataFrames
parking_lots_gdf = gpd.read_file('SFMTA_Managed_Off-street_Parking_20240404.csv')
crime_gdf = gpd.read_file('Police_Department_Incidents_-_Previous_Year__2016_.csv')

# Convert latitude and longitude columns to Point objects
parking_lots_gdf['geometry'] = parking_lots_gdf.apply(lambda row: Point(row['MAIN_ENTRANCE_LONG'], row['MAIN_ENTRANCE_LAT']), axis=1)
crime_gdf['geometry'] = crime_gdf.apply(lambda row: Point(row['Y'], row['X']), axis=1)

crime_gdf
```

```python
# Define a function to check if a crime has taken place in within 1 mile of a parking lot
def within_one_mile(parking_lot, crime_spots):
    return crime_spots[crime_spots.geometry.apply(lambda crime_spot:distance((parking_lot['MAIN_ENTRANCE_LAT'], parking_lot['MAIN_ENTRANCE_LONG']), (crime_spot.x, crime_spot.y)).miles <= 1)]

# Perform spatial join to get restaurants within 1 mile of each parking lot
joined_data = gpd.GeoDataFrame()
for idx, parking_lot in parking_lots_gdf.iterrows():
    crim_spot_within_one_mile = within_one_mile(parking_lot, crime_gdf)
    crim_spot_within_one_mile['parking_lot_id'] = parking_lot['OBJECTID']  # Add parking lot ID to identify the source
    joined_data = pd.concat([joined_data, crim_spot_within_one_mile], ignore_index=True)


# Save or display the joined data
#joined_data.to_file('joined_data.geojson', driver='GeoJSON')  # Save to GeoJSON file
print(joined_data)
```

```python
# Convert the GeoDataFrame to a pandas DataFrame
df = pd.DataFrame(joined_data.drop(columns=['geometry']))

print(df)
```
